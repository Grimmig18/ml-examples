```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from scipy.spatial import distance

iris = datasets.load_iris()

iris_data = iris.data
iris_labels = iris.target
```

```python
from scipy.cluster.vq import vq, kmeans, whiten
import matplotlib.pyplot as plt
```

```python
whitened = whiten(iris_data)
print(np.array(whitened).shape)

book = np.array((whitened[0],whitened[50],whitened[100]))
print(book)

codebook, dist = kmeans(whitened, book, iter=100)
```

```python
fig = plt.figure(dpi=200)
ax = fig.add_subplot(111, projection='3d')

xs = [p[3] for p in whitened]
ys = [p[2] for p in whitened] 
zs = [p[1] for p in whitened]
h  = [p[0] for p in whitened]

ax.scatter(xs, ys, zs, c=h, cmap=plt.hot())

xs = [p[3] for p in codebook]
ys = [p[2] for p in codebook] 
zs = [p[1] for p in codebook]
h  = [p[0] for p in codebook]

img = ax.scatter(xs, ys, zs, c='b')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

ax.set_xlim([0, 3])
ax.set_ylim([0, 4])
ax.set_zlim([4, 10])

fig.colorbar(img)

print(codebook)
```
